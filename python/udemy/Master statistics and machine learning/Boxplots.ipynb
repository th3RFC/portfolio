## Boxplots

In this notebook, we will be experimenting with generating boxplots.

```python
# Import libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

First, we must create data to use for our boxplots:

```python
# Create data for barplot

# Data sizes
m = 30 
n = 6

data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30 * np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

Having created the data, our first boxplot is:

```python
# Boxplot
plt.boxplot(data)
plt.show;
```

```python
# Boxplot transposed
plt.boxplot(data.T)
plt.show;
```

We can also create boxplots using **Seaborn**:

```python
# Boxplot with seaborn
sns.boxplot(data=data, orient='v')
plt.show;
```

```python
# Using a pandas data frame
df = pd.DataFrame(data, columns=['zero','one','two','three','four','five'])
sns.boxplot(data=df, orient='h')
plt.show;
```
