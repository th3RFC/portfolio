```python
#@title Download Performance as MIDI
#@markdown Download generated performance as MIDI (optional).
from transformers import pipeline

# Specify the model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
sentiment_pipeline = pipeline("sentiment-analysis", model=model_name)
```

```python
# Sample texts for sentiment analysis
sample_texts = [
    "I love sunny days in the city!",
    "Traffic jams are the worst.",
    "Had a mediocre lunch; not bad, but not great either."
]
```

```python
# Analyze sentiment
results = sentiment_pipeline(sample_texts)

# Display the results
for text, sentiment in zip(sample_texts, results):
    print(f"Text: {text}\nSentiment: {sentiment}\n")
```
